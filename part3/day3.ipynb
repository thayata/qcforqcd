# 情報スクランブリング

局所的に入力した量子情報が, ダイナミクスにより量子多体系全体に拡散し, 多体のエンタングルメントに情報がエンコードされることで, 局所的な測定からは情報を復元できなくなる現象を情報スクランブリングという

* 非局所的に情報が保持されているのであって, 完全に喪失したわけではないので, 開放系におけるデコヒーレンスとは区別できる現象である

* 初期状態の情報を喪失する点において熱化と似ているが, 熱化が局所的な物理量が統計平均で記述されるかどうか(初期状態を忘れるかどうか)を問題にするのに対して, 情報スクランブリングは熱化で喪失した情報が非局所的な自由度に流れていく過程(早さやその程度)を問題にする. その点において情報スクランブリングは量子多体系の非平衡ダイナミクスをより詳細に理解するのに有用な物理現象である

* 実際, 情報スクランブリングを定量化するための物理量の一つに非時間順序積相関関数(out of time-ordered correlator; OTOC)があるが, OTOCは時間に関するある種の高次のモーメントであり, 高次のモーメントを測ることで状態をより精密に特徴付けていると理解することができる


スクランブリングのダイナミクスを抽象的に図示すると以下のようになる

熱化までの時間は局所的な量, 例えば, 系の平均的なエネルギー密度などで決まるのに対して, スクランブリングまでの時間$t_*$は典型的に対象の局所系が系全体と強く相互作用するまでの時間で決まり, 場の理論は局所相互作用する理論なので, $t_*$は系の大きさに強く依存する

ブラックホール（やSachdev-Yee-Kitaev模型)は最も早い時間でスクランブリング示す物理系であると考えられているが, これらの系ではスクランブリング時間は系のサイズのlog程度であるとされている (通常はpoly程度)

スクランブリングの必要条件に, スクランブリングを起こすユニタリ発展から状態をサンプリングすると, ランダムサンプリングする(Haar random unitaryからサンプルする)のと見分けがつかないという性質がある

そのため, スクランブリングのダイナミクスをテンソルネットワーク等を用いて効率的に（精度良く近似）計算することは難しいと考えられている

情報スクランブリングを特徴づけるいくつかの等価な物理量がある

それらを一つ一つ紹介していく

# Operator growthによる定義

局所的な情報の入力を局所的な励起と読み換えると, 情報スクランブリングは励起の伝搬するダイナミクスから理解できるだろう

ある局所的な演算子$W$による励起が時間発展とともに系全体に拡散すれば, ハイゼンベルグ演算子$W(t)$は非局所な演算子になり, 初期時刻に＄[W,V]=0$であった別の場所に定義された局所的な演算子$V$と交換しなくなる: $[W(t),V]\neq 0$

非可換性の程度(ハイゼンベルグ演算子の台の成長具合)を観測することで, 情報スクランブリングを定量化するのがOperator growthに基づく定義である


### Out-of-time-ordered correlator (OTOC)の導入

単純には$\langle\psi(0)|[W(t),V]|\psi(0)\rangle$を測れば良さそうである

しかしながら, 熱化する系では$e^{-{\rm i}Ht}V|\psi(0)\rangle$と$e^{-{\rm i}Ht}|\psi(0)\rangle$の間の違いは分からなくなってしまうので,

$\langle\psi(0)|[W(t),V]|\psi(0)\rangle$はいつでも$\sim0$であまり良い指標ではない



これを解決するために高次のモーメントを考える: $\langle\psi(0)|[W(t),V]^\dagger[W(t),V]|\psi(0)\rangle$

$W,V$としてエルミートかつユニタリな演算子（パウリストリングなど）を採用しても一般性をそれほど損なわないだろう

その場合, $\langle\psi(0)|[W(t),V]^\dagger[W(t),V]|\psi(0)\rangle=2-2{\rm Re}\left[\langle \psi(0)|W(t)VW(t)V|\psi(0)\rangle\right]$となる

ここで出てくる$\langle \psi(0)|W(t)VW(t)V|\psi(0)\rangle$がOTOCである

スクランブリングは$2k$次のOTOC (上の式は$k=1$)が減衰し, 平均OTOCが$\sim \frac{1}{d_{V}^{2k}}$のように振る舞う, より正確にはHaar random unitaryに対して平均OTOCを計算した結果に一致することをもって定義される ($d\gg d_W\gg d_V$を仮定した)

スクランブリング時間はOTOCが$\sim \frac{1}{d_V^{2k}}$に到達するまでの時間である

何次までHaar random unitaryに一致するかで階層性がある


### ラフな説明

スクランブリング(operator growth)$\rightarrow{\rm OTOC}\sim0$は次のように理解できる

${\rm tr}\left[Z_n(t)Z_1Z_n(t)Z_1\right]$を計算するとしよう

$Z_n(t)=\sum_{\mu_1,\ldots, \mu_N}a_{\mu_1\cdots \mu_N}(t)\sigma_{\mu_1}\otimes\cdots\otimes\sigma_{\mu_N}$

$Z_1$は$\sigma_{\mu_1}=1,Z$と可換で, $\sigma_{\mu_1}=X,Y$と反可換なので, trに寄与するのは

${\rm tr}\left[Z_n(t)Z_1Z_n(t)Z_1\right]=\sum_{\mu_2,\ldots, \mu_N}|a_{0\mu_2\cdots \mu_N}(t)|^2+\sum_{\mu_2,\ldots, \mu_N}|a_{Z\mu_2\cdots \mu_N}(t)|^2-\sum_{\mu_2,\ldots, \mu_N}|a_{X\mu_2\cdots \mu_N}(t)|^2-\sum_{\mu_2,\ldots, \mu_N}|a_{Y\mu_2\cdots \mu_N}(t)|^2$

スクランブリングが起きると高次の$\sigma_{\mu_1}\otimes\cdots\otimes\sigma_{\mu_N}$が一様に$Z_n(t)$へ寄与するので, 高次の$|a_{\mu_1\mu_2\cdots \mu_N}(t)|^2$は大体同じ値$\rightarrow{\rm tr}\left[Z_n(t)Z_1Z_n(t)Z_1\right]\sim0$



# Unitary $k$-designによる定義

スクランブリングの単純化した説明に, 対象のunitary circuit (例えば$U(t)={\rm e}^{-{\rm i}Ht}$) がHaar random unitaryと区別がつかない, もしくは$U(t)$で発展させて物理量の期待値を計算すると完全混合状態と一致するという説明がある

この説明を定量化したものがunitary k-designに基づくスクランブリングの定義である

定義を具体的にするためにハミルトニアンによる時間発展$U(t)={\rm e}^{-{\rm i}Ht}$を考え, $U(t)^m$ ($m=0,1,2,\cdots,K$)によって生成されるユニタリ演算子(量子チャンネル)のアンサンブル${\cal E}$を考える

${\cal H}^{\otimes k}$上の任意の量子状態$\rho$に対して,

$\sum_{U\in {\cal E}}U^{\otimes k}\rho U^{\otimes k}=\int_{\rm Haar}dU  U^{\otimes k}\rho U^{\otimes k}$

が成り立つとき${\cal E}$はunitary $k$-designであるという

unitary k-design $\rightarrow$ $2k$次のOTOCがHaar randomと一致


### state $k$-design


スクランブリングを示す量子回路はtraceからのランダムサンプリングに利用することができる

この場合は状態のアンサンブルを生成したいので, 適当な初期状態$|\psi_0\rangle$を固定して, $|\psi\rangle=U(t)^m|\psi_0\rangle$ ($m=0,1,2,\cdots,K$)によって生成される状態のアンサンブル${\cal E}$を考える

任意の演算子$O$に対して

${\mathbf{E}}_{|\psi\rangle\in {\cal E}}\left[|\psi\rangle^{\otimes k}\langle\psi|^{\otimes k}O\right]={\mathbf{E}}_{|\psi\rangle\in {\rm Haar}}\left[|\psi\rangle^{\otimes k}\langle\psi|^{\otimes k}O\right]$

が成り立つとき${\cal E}$はstate $k$-designであるという

実用上の問題でstate $2$-design以上が出ることは知る限りない

ギブス状態での統計平均 (trace)をstate $2$-designを用いて計算することを考える

state $2$-designからサンプリングすることで得られる物理量の推測値の分散は系のサイズに対して指数関数的に小さくなる:

${\rm Prob}_{\psi\in{\rm 2-design}}[|\langle O\rangle^{\rm est}-\langle O\rangle|\leq\varepsilon]\geq 1-{\cal O}\left({\rm e}^{-cN}/\varepsilon^2\right)$

分散が小さくなることを保証するのにstate $2$-designが必要となる

# 相互情報量による定義

ユニタリ回路 $U$の局所的なqubits $A$に量子状態（量子情報）を入力するとする

量子情報の拡散とは, $A$の情報が$C$もしくは$D$のどちらかだけでは得られないこと意味するので, 式で書けば,

情報スクランブリングは$I(A:CD)-I(A:C)-I(A:D)$が最大化した状態である 

($I(X:Y)=S_X+S_Y-S_{XY}$は$X$と$Y$の相互情報量, $S$はエンタングルメントエントロピー)

一方で$3$点の情報量 $I_3(A:C:D)=S_A+S_C+S_D-S_{AC}-S_{AD}-S_{CD}+S_{ACD}=-\left(I(A:CD)-I(A:C)-I(A:D)\right)$なので,
情報スクランブリングは$I_3(A:C:D)$が最小化した状態である

inとoutのエンタングルメントエントロピーはEPR pairsの挿入を介して測るものとして理解する

$|{\rm EPR}\rangle=\frac{1}{\sqrt{d}}\sum_{i=1}^d|i\rangle\otimes|i\rangle$

### OTOCとの関係

平均OTOC $\frac{1}{d_A^2d_D^2}\sum_{O_A,O_C}{\rm tr}\left[O_A(t)O_DO_A(t)O_D \right]$を計算する

$d_A$は部分系$A$のヒルベルト空間の次元

次の恒等式が成り立つ

$\frac{1}{d_A}\sum_{O_A}O_A O O_A=I_A\otimes {\rm tr_A}O$

例えば1qubitの場合は$\frac{1}{2}\sum_{\mu}\sigma_\mu \left(\sum_{\nu}o_\nu\sigma_\nu\right)\sigma_\mu=\frac{1}{2}\sum_{\nu}o_\nu\sum_{\mu}\sigma_\mu \sigma_\nu\sigma_\mu$

$X,Y,Z$と$\sigma_\mu$の間には可換と非可換のペアが$2$つずつあるので

例えば$\sum_{\mu}\sigma_\mu X\sigma_\mu=IXI+XXX+YXY+ZXZ=0$となり, Iの成分($\nu=0$)のみが残る

したがって$\frac{1}{2}\sum_{\mu}\sigma_\mu \sum_{\nu}o_\nu\sigma_\nu\sigma_\mu=2o_0={\rm tr}O$

ダイアグラム表示すると

足が閉じているところはtrを意味する

この恒等式を使って平均OTOCを書き換える

$\frac{1}{d_A^2d_D^2}\sum_{O_A,O_C}{\rm tr}\left[O_A(t)O_DO_A(t)O_D \right]$のダイアグラムは



$=\frac{1}{d_Ad_D} {\rm tr}\rho_{AC}^2=2^{-N_A-N_D-S^{(2)}_{AC}}$

OTOCの減衰はACもしくはAD間のエンタングルメントエントロピーの増大と対応している

つまり$I(A:C)=S_A+S_C-S_{AC}$もしくは$I(A:D)$の減少と対応している

完全にスクランブリングされた状態では$I(A:C)\sim I(A:D)\sim 0$, $I_3(A:C:D)\sim -I(A:CD)\sim -2N_A$


# Hayden-Preskillの思考実験

$U$がscramblingを示すとき, $|\psi_A\rangle$は少数の$D$から再構築可能

$\leftrightarrow$テレポーテーションの成功確率が$1$に近いとき, $U$はscramblingを示す

RR'がEPR pairsかどうかで$U$がscramblingを示すかどうかを測定できる

# Yoshida-Kitaevのdecoding protocol (non-deterministic)

* $AB$を$U$, $A'B'$を$U^*$で発展させる

* $DD'$に対して$|\text{EPR}\rangle$に射影測定を行う

* 射影測定に成功した場合のみ, $RR'$について$|\text{EPR}\rangle$に射影測定を行う

射影測定に成功する確率$P_{\rm EPR}$

$P_\text{EPR}$は$A$と$D$のあいだの平均OTOCに一致する

RR'がEPR pairsである確率$F_{\rm EPR}$

平均OTOCが減衰するとき, $F_{\rm EPR}$が$1$に近い

量子回路図を書くと

量子テレポーテーションの多体系への拡張と見なすことができる

まとめると

* Operator growth: high-weight Pauli stringsの寄与が顕著になる

* OTOCの減衰

* Unitary $k$-design: Haar random unitaryとどれぐらい区別がつかないか

* 3点相互情報量が負に最大化する (bipartite informationがゼロになる)

* Hayden-Preskill decoding protocolの成功確率 (~多体の量子テレポーテーション)

基本的に等価な物理を見ているが, 開放系だったりノイズのある量子回路では異なる振る舞いをすることがあることも知られている

## Haar random unitaryの場合の平均OTOCの計算

Haar random 積分公式

N qubit系であれば, d= 2^N. $U$はd^2次元のユニタリ行列

$$
\int {\cal D}U =1
$$


$$
 \int {\cal D}U U_{i_1j_1}U^\ast_{i_2j_2}
 =\frac{\delta_{i_1i_2}\delta_{j_2j_1}}{d}
$$

$$
 \int{\cal D}U  U_{i_1j_1}U_{i_2j_2}U^\ast_{i_3j_3}U^\ast_{i_4j_4}
 =\frac{\delta_{i_1i_3}\delta_{i_2i_4}\delta_{j_1j_3}\delta_{j_2j_4} 
 + \delta_{i_1i_4}\delta_{i_2i_3}\delta_{j_1j_4}\delta_{j_2j_3}}{d^2-1}
 -\frac{\delta_{i_1i_3}\delta_{i_2i_4}\delta_{j_1j_4}\delta_{j_2j_3} 
 + \delta_{i_1i_4}\delta_{i_2i_3}\delta_{j_1j_3}\delta_{j_2j_4}}{d(d^2-1)}.
$$

$$
P_\text{EPR} 
    =\frac{1}{d_A^2d_Bd_D}U_{a_1b_1c_1d_1}U_{a_2b_2c_2d_2}U^\ast_{a_1b_2c_1d_2}U^\ast_{a_2b_1c_2d_1}.
$$

$$
 \int{\cal D}U P_\text{EPR}
 =\frac{1}{d_A^2d_Bd_D}\int\int{\cal D}U  U_{a_1b_1c_1d_1}U_{a_2b_2c_2d_2}U^\ast_{a_1b_2c_1d_2}U^\ast_{a_2b_1c_2d_1}
 $$


 $$
 =\frac{1}{d_A^2d_Bd_D}\Big[\frac{\delta_{a_1a_1}\delta_{b_1b_2}\delta_{a_2a_2}\delta_{b_2b_1}\ \delta_{c_1c_1}\delta_{d_1d_2}\delta_{c_2c_2}\delta_{d_2d_1} 
 + \delta_{a_1a_2}\delta_{b_1b_1}\delta_{a_2a_1}\delta_{b_2b_2}\ \delta_{c_1c_2}\delta_{d_1d_1}\delta_{c_2c_1}\delta_{d_2d_2}}{d^2-1}
 $$

 $$
-\frac{\delta_{a_1a_1}\delta_{b_1b_2}\delta_{a_2a_2}\delta_{b_2b_1}\ \delta_{c_1c_2}\delta_{d_1d_1}\delta_{c_2c_1}\delta_{d_2d_2} 
 + \delta_{a_1a_2}\delta_{b_1b_1}\delta_{a_2a_1}\delta_{b_2b_2}\ \delta_{c_1c_1}\delta_{d_1d_2}\delta_{c_2c_2}\delta_{d_2d_1}}{d(d^2-1)}\Big]
$$

$$
 =\frac{1}{d_A^2d_Bd_D}\Big[\frac{d_A^2d_Bd_C^2d_D + d_Ad_B^2d_Cd_D^2}{d^2-1}
 -\frac{d_A^2d_Bd_Cd_D^2 + d_Ad_B^2d_C^2d_D}{d(d^2-1)}\Big]
$$

$$
 =\frac{1}{d^2-1}\left[d_B^2 + d_C^2
 -\frac{d_C^2}{d_A^2}-1\right]\approx \frac{1}{d_A^2}+\frac{1}{d_D^2}-\frac{1}{d_A^2}\frac{1}{d_D^2}\approx \frac{1}{d_A^2}
$$

$$F_\text{EPR}=\frac{1}{d_A^2P_\text{EPR}}\approx \frac{1}{1+\frac{d_A^2}{d_D^2}}\approx 1$$

# OTOCの測り方

二つの領域$A$, $D$上に定義された演算子$O_A$, $O_D$に関するOTOC=$\langle\psi|O_D^\dagger(t)O^\dagger_AO_D(t)O_A|\psi\rangle$は次のHadamard testを用いて測定できる

$|+\rangle\otimes|\psi\rangle$

C-$O_A\rightarrow\frac{1}{\sqrt{2}}|0\rangle\otimes|\psi\rangle+\frac{1}{\sqrt{2}}|1\rangle\otimes O_A|\psi\rangle$

$O_D(t)\rightarrow\frac{1}{\sqrt{2}}|0\rangle\otimes O_D(t)|\psi\rangle+\frac{1}{\sqrt{2}}|1\rangle\otimes O_D(t)O_A|\psi\rangle$

C-$O_A^\dagger\rightarrow\frac{1}{\sqrt{2}}|0\rangle\otimes O_D(t)|\psi\rangle+\frac{1}{\sqrt{2}}|1\rangle\otimes O_A^\dagger O_D(t)O_A|\psi\rangle$

期待値$\langle X\rangle$は

$\langle X\rangle=\frac{1}{2}\left(\langle0|\otimes \langle\psi|O_D^\dagger(t)+\langle1|\otimes \langle\psi|O_A^\dagger O_D^\dagger(t)O_A\right)X\left(|0\rangle\otimes O_D(t)|\psi\rangle+|1\rangle\otimes O_A^\dagger O_D(t)O_A|\psi\rangle\right)$

$\hspace{2em}=\frac{1}{2}\left(\langle\psi|O_D^\dagger(t)O_A^\dagger O_D(t)O_A|\psi\rangle+\langle\psi|O_A^\dagger O_D^\dagger(t)O_AO_D(t)|\psi\rangle \right)={\rm Re}\langle\psi|O_D^\dagger(t)O_A^\dagger O_D(t)O_A|\psi\rangle$

期待値$\langle Y\rangle$は

$\langle Y\rangle=\frac{1}{2}\left(\langle0|\otimes \langle\psi|O_D^\dagger(t)+\langle1|\otimes \langle\psi|O_A^\dagger O_D^\dagger(t)O_A\right)Y\left(|0\rangle\otimes O_D(t)|\psi\rangle+|1\rangle\otimes O_A^\dagger O_D(t)O_A|\psi\rangle\right)$

$\hspace{2em}=\frac{1}{2}\left((-{\rm i})\langle\psi|O_D^\dagger(t)O_A^\dagger O_D(t)O_A|\psi\rangle+{\rm i}\langle\psi|O_A^\dagger O_D^\dagger(t)O_AO_D(t)|\psi\rangle \right)={\rm Im}\langle\psi|O_D^\dagger(t)O_A^\dagger O_D(t)O_A|\psi\rangle$
